# Embeddings-based search engine
This document explores how to implement Embeddings-based search engine. I will first explore libraries and decide which ones I will use, and then based on this scribt built my search engine

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
sentences = ["This is an example sentence", "Each sentence is converted"]
embeddings = model.encode(sentences)
print(embeddings.shape)
print(embeddings)
```

This is implemented from hugging face example.

I will implement in the following lines faiss. I choose this over other libraries because it is the easiest to implement and works totally fine for prototype/research phase.

```python
import numpy as np
import faiss

d = embeddings.shape[1]  # embedding dimension
index = faiss.IndexFlatIP(d)  # IP = inner product

index.add(embeddings)  # add vectors to the index
query = "This is a test sentence to give example"
query_emb = model.encode([query], normalize_embeddings=True)
k = 2
scores, indices = index.search(query_emb, k)

print("\nResults:")
for rank, idx in enumerate(indices[0]):
    print(f"{rank+1}. {sentences[idx]} (score={scores[0][rank]:.4f})")
```

This is the simplest implementation of Bi-encoder which is expected to work well but less accurate then cross-encoder, while cross-encoder might be slower. For now I will stick with bi-encoder for simplicity and since this is ment to be more of an MVP. But in fruther implementations it could include hybrid approach where on top of basic bi-encoder, cross-encoder could be built another that fruther fine-tunes top k retrievals.
